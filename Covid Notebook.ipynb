```python
import pandas as pd
```

```python
confrimed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
confrimed_df
```

```python
new_df = confrimed_df.melt(id_vars=['Country/Region','Province/State','Lat','Long'])
```

```python
new_df[new_df['Country/Region']=='India'].head()
```

```python
temp_val = 0
def dailyCaseClac(x):
    global temp_val
    currentVal = x - temp_val
    temp_val = x
    return int(currentVal)
```

```python
new_df['Daily_Case'] = new_df[new_df['Country/Region'] == 'India']['value'].apply(lambda x: dailyCaseClac(x))
```
